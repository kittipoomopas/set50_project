```python
!pip install flask flask-cors requests
from flask import Flask, jsonify
from flask_cors import CORS
import requests

app = Flask(__name__)
CORS(app) # อนุญาตให้ไฟล์ HTML โลคอลดึงข้อมูลจาก API นี้ได้

symbols = [
    "SET:ADVANC","SET:AOT","SET:AWC","SET:BANPU","SET:BBL",
    "SET:BDMS","SET:BEM","SET:BGRIM","SET:BH","SET:BJC",
    "SET:BTS","SET:CBG","SET:CENTEL","SET:CPALL","SET:CPF",
    "SET:CPN","SET:CRC","SET:DELTA","SET:EA","SET:EGCO",
    "SET:PTT","SET:SCB","SET:SCC" # ย่อลงเพื่อให้ดูง่าย
]

url = "https://scanner.tradingview.com/thailand/scan"

payload = {
    "symbols": {"tickers": symbols, "query": {"types": []}},
    "columns": ["close", "open", "high", "low", "volume", "change", "change_abs"]
}

@app.route('/api/stocks', methods=['GET'])
def get_stocks():
    r = requests.post(url, json=payload)
    data = r.json()
    
    stock_list = []
    for stock in data["data"]:
        stock_list.append({
            "symbol": stock["s"].replace("SET:", ""),
            "price": stock["d"][0],
            "change": stock["d"][5]
        })
        
    return jsonify(stock_list)

if __name__ == '__main__':
    print("🚀 สตาร์ทเซิร์ฟเวอร์ที่ http://localhost:5000")
    app.run(debug=True, use_reloader=False, port=5000)

import feedparser
import time
from flask import Flask, jsonify
from flask_cors import CORS
import requests

# ... (โค้ด Flask และ API หุ้นตัวเดิมของคุณ) ...

@app.route('/api/news', methods=['GET'])
def get_news():
    # ใช้ RSS Feed ข่าวการเงิน/หุ้น ของสำนักข่าว (ในที่นี้ใช้ประชาชาติธุรกิจเป็นตัวอย่าง)
    rss_url = "https://www.prachachat.net/finance/feed"
    feed = feedparser.parse(rss_url)
    
    news_list = []
    # ดึงมาแค่ 5 ข่าวล่าสุด เพื่อให้พอดีกับกล่องในหน้า Dashboard
    for entry in feed.entries[:5]:
        # จัดฟอร์แมตเวลาดึงมาแค่ ชั่วโมง:นาที
        if hasattr(entry, 'published_parsed'):
            # เวลาใน RSS มักเป็นเวลาสากล (GMT) +7 ชั่วโมงเพื่อให้เป็นเวลาไทย (ถ้าจำเป็น)
            time_str = time.strftime("%H:%M", entry.published_parsed) 
        else:
            time_str = "New"
            
        news_list.append({
            "time": time_str,
            "title": entry.title,
            "link": entry.link # ส่งลิงก์ข่าวจริงไปให้หน้าเว็บด้วย
        })
        
    return jsonify(news_list)

# ... (โค้ด app.run ด้านล่างสุด) ...
```
